```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

data = {
    'EmailText': [
        'Win money now',
        'Limited offer claim prize',
        'Meeting at 10 am',
        'Project submission tomorrow',
        'Congratulations you won lottery',
        'Call me when free',
        'Free vacation offer',
        'Important bank update',
        'Earn dollars quickly',
        'Lets have lunch today',
        'Claim your free reward',
        'Your invoice is attached',
        'Cheap loans available',
        'Team meeting schedule',
        'Exclusive discount offer'
    ],

    'Label': [
        'Spam',
        'Spam',
        'Ham',
        'Ham',
        'Spam',
        'Ham',
        'Spam',
        'Ham',
        'Spam',
        'Ham',
        'Spam',
        'Ham',
        'Spam',
        'Ham',
        'Spam'
    ]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

df['Label'] = df['Label'].map({
    'Ham': 0,
    'Spam': 1
})

X = df['EmailText']
y = df['Label']

vectorizer = CountVectorizer()

X_vectorized = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.3,
    random_state=42
)

model = MultinomialNB()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\n================================")
print("Confusion Matrix")
print("================================")

cm = confusion_matrix(y_test, y_pred)
print(cm)

print("\n================================")
print("Performance Metrics")
print("================================")

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)

print("\n================================")
print("Classification Report")
print("================================")

print(classification_report(y_test, y_pred))

sample_email = ["Congratulations! Claim your free cash reward now"]

sample_vector = vectorizer.transform(sample_email)

prediction = model.predict(sample_vector)

print("\n================================")
print("Custom Email Prediction")
print("================================")

if prediction[0] == 1:
    print("Spam Email")
else:
    print("Not Spam (Ham)")
```

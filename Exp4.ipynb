```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

data = load_iris()

X = data.data
y = data.target

feature_names = data.feature_names

df = pd.DataFrame(X, columns=feature_names)

print("Dataset Head:\n")
print(df.head())

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])

print("\nPCA Reduced Features:\n")
print(pca_df.head())


lda = LinearDiscriminantAnalysis(n_components=2)

X_lda = lda.fit_transform(X_scaled, y)

lda_df = pd.DataFrame(X_lda, columns=['LDA1', 'LDA2'])

print("\nLDA Reduced Features:\n")
print(lda_df.head())


X_train_pca, X_test_pca, y_train, y_test = train_test_split(
    X_pca, y,
    test_size=0.2,
    random_state=42
)

model_pca = LogisticRegression()

model_pca.fit(X_train_pca, y_train)

y_pred_pca = model_pca.predict(X_test_pca)

pca_accuracy = accuracy_score(y_test, y_pred_pca)

print("\nPCA Classification Accuracy:")
print(pca_accuracy)


X_train_lda, X_test_lda, y_train, y_test = train_test_split(
    X_lda, y,
    test_size=0.2,
    random_state=42
)

model_lda = LogisticRegression()

model_lda.fit(X_train_lda, y_train)

y_pred_lda = model_lda.predict(X_test_lda)

lda_accuracy = accuracy_score(y_test, y_pred_lda)

print("\nLDA Classification Accuracy:")
print(lda_accuracy)


plt.figure(figsize=(6,5))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=y
)

plt.title("PCA 2D Projection")
plt.xlabel("PCA1")
plt.ylabel("PCA2")

plt.show()


plt.figure(figsize=(6,5))

plt.scatter(
    X_lda[:,0],
    X_lda[:,1],
    c=y
)

plt.title("LDA 2D Projection")
plt.xlabel("LDA1")
plt.ylabel("LDA2")

plt.show()


print("\nAccuracy Comparison")
print("PCA Accuracy :", pca_accuracy)
print("LDA Accuracy :", lda_accuracy)

print("\nComments:")
print("PCA focuses on maximum variance.")
print("LDA focuses on maximum class separation.")
print("LDA usually gives better class separation than PCA.")
```
